```python
import random
import networkx as nx
```

```python
G=  nx.Graph()
#G= nx.DiGraph() #directed graph

city_set= ['a','b','c','d','e']
for i in city_set:
  G.add_node(i)
print(type(G.nodes()))

costs= []
value=100
while(value<=2000):
  costs.append(value)
  value+=100

while (G.number_of_edges()<5):
  c1= random.choice(list(G.nodes()))
  c2= random.choice(list(G.nodes()))
  
  if (c1!=c2 and G.has_edge(c1,c2)==0):
    w= random.choice(costs)
    G.add_edge(c1,c2,weight=w)

pos= nx.circular_layout(G)
nx.draw(G,pos, with_labels=1)
#nx.draw_networkx_edge_labels(G,pos)
plt.show()
```

```python
print (nx.is_connected(G))
```

```python
#checks if theres a path between every pair of node i.e. if 2 nodes are connected 
for u in G.nodes():
  for v in G.nodes():
    print(u,v,nx.has_path(G,u,v))
```

```python
def create_network(city_set,costs,num_edges):
  G=  nx.Graph()
  for i in city_set:
    G.add_node(i)

  while (G.number_of_edges()<num_edges):
    c1= random.choice(list(G.nodes()))
    c2= random.choice(list(G.nodes()))
    
    if (c1!=c2 and G.has_edge(c1,c2)==0):
      w= random.choice(costs)
      G.add_edge(c1,c2,weight=w)
      
  return G
```

```python
city_set= ['a','b','c','d','e']
costs= []
value=100
while(value<=2000):
  costs.append(value)
  value+=100
```

```python
G= create_network(city_set,costs,5)
nx.draw(G, with_labels=1)
plt.show()

print(nx.dijkstra_path(G,'a','d'))
print(nx.dijkstra_path_length(G,'a','d'))
```

```python
G= create_network(city_set,costs,0)
nx.draw(G, with_labels=1)
plt.show()

print(nx.dijkstra_path(G,'a','d'))
print(nx.dijkstra_path_length(G,'a','d'))
```

```python
def add_random_edge(G,costs):
  c1= random.choice(list(G.nodes()))
  c2= random.choice(list(G.nodes()))
    
  if (c1!=c2 and G.has_edge(c1,c2)==0):
    w= random.choice(costs)
    G.add_edge(c1,c2,weight=w)
```

```python
G= create_network(city_set,costs,0)
nx.draw(G, with_labels=1)
plt.show()

x=[0]
y=[l]

for t in range(1,10000):
  add_random_edge(G,costs)
  x.append(t)
  try:
    l= nx.dijkstra_path_length(G,'a','d')
    y.append(l)
    break
  except:
    l=10000
    y.append(10000)
    print('no path')

nx.draw(G, with_labels=1)
plt.show()

plt.xlabel('Time')
plt.ylabel('Cost')
plt.title('Change in cost with increase in edges')
plt.plot(x,y)
plt.show()
```
